Data cleaning and preperation

```python
#column:- Unnamed: 0 has a simple index number nothing else so drop it
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
df.head(5)
```

```python
#in column 'competitors' there is a -1 so it need to be replace with 'unknown'
df.replace('-1',pd.NA,inplace=True)#unknown
df
```

```python
#in column 'Salary Estimate' no need for analysis of  (Glassdoor est.) so remove it
df['Salary Estimate'] = df['Salary Estimate'].str.replace('(Glassdoor est.)', '')
df.head()
```

```python
df['Salary Estimate'] = df['Salary Estimate'].str.replace('K', '000')
df.head(2)
```

```python
df['Salary Estimate'] = df['Salary Estimate'].str.replace('(Employer est.)', '')
```

```python
df['Salary Estimate'] = df['Salary Estimate'].str.replace('$', '')
df.head()
```

```python
df['Salary Estimate'].unique()
```

```python
#covertting the column 'Founded' to datetime format
df['Founded']=pd.to_datetime(df['Founded'],format='%Y',errors='coerce')
df
```

```python
#we have already rating so drop rating present in 'Company Name'
df['Company Name']=df['Company Name'].str.split('\n').str.get(0)
df
#df['Company Name']
```

```python
#check null value after data preparation
#calulate the percentage of null value
null_value_percentage = (df.isnull().sum()/len(df))*100
null_value_percentage
```

```python
#The coulmns have less than 30 percent null value then replace with its aggregate.
```

```python
df['Salary Estimate'].fillna(df['Salary Estimate'].mode()[0],inplace=True)
```

```python
df['Headquarters'].fillna(df['Headquarters'].mode()[0],inplace=True)
```

```python
df['Size'].fillna(df['Size'].mode()[0],inplace=True)
```

```python
df['Founded'].fillna(df['Founded'].mode()[0],inplace=True)
```

```python
df['Type of ownership'].fillna(df['Type of ownership'].mode()[0],inplace=True)
```

```python
df['Industry'].fillna(df['Industry'].mode()[0],inplace=True)
```

```python
df['Sector'].fillna(df['Sector'].mode()[0],inplace=True)
```

```python
df['Revenue'].fillna(df['Revenue'].mode()[0],inplace=True)
```
